**2. Statistics and Probability in Data Science**

---



Tujuan:
* Memahami konsep dasar statistik dan probabilitas yang digunakan dalam data sains.
* Mampu menghitung statistik dasar dan probabilitas sederhana menggunakan Python.

Materi:
* Mean, Median, Mode
* Variance dan Standard Deviation
* Distribusi Probabilitas

```python
import numpy as np
from scipy import stats

# Data sample
data = [12, 15, 12, 15, 16, 18, 19, 20, 22, 25]

# Mean, Median, Mode
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode.mode)

# Variance dan Standard Deviation
variance = np.var(data)
std_dev = np.std(data)

print("Variance:", variance)
print("Standard Deviation:", std_dev)

# Distribusi Probabilitas: Binomial
n = 10  # Jumlah percobaan
p = 0.5  # Probabilitas sukses
k = 6  # Jumlah sukses

binom_prob = stats.binom.pmf(k, n, p)
print(f"Probabilitas mendapatkan {k} sukses dalam {n} percobaan:", binom_prob)
```

**2. Statistics and Probability in Data Science (*Jika kita mengimpor data dari file CSV*)**

---



```python
import pandas as pd
from scipy import stats

# Mengimpor data dari file CSV
data_df = pd.read_csv('/content/data.csv')

# Mengambil kolom data sebagai array
data = data_df['nilai'].values

# Menghitung Mean, Median, dan Mode
mean = data_df['nilai'].mean()
median = data_df['nilai'].median()
mode = stats.mode(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode.mode)

# Menghitung Variance dan Standard Deviation
variance = data_df['nilai'].var()
std_dev = data_df['nilai'].std()

print("Variance:", variance)
print("Standard Deviation:", std_dev)

# Distribusi Probabilitas: Binomial
n = 10  # Jumlah percobaan
p = 0.5  # Probabilitas sukses
k = 6  # Jumlah sukses

binom_prob = stats.binom.pmf(k, n, p)
print(f"Probabilitas mendapatkan {k} sukses dalam {n} percobaan:", binom_prob)
```

**3. Machine Learning (*Linear Regression*)**

---



Tujuan:
* Memahami konsep dasar machine learning.
* Mampu mengimplementasikan model machine learning sederhana menggunakan Python.

Materi:
* Supervised Learning: Linear Regression
* Unsupervised Learning: K-Means Clustering


```python
# Linear Regression
from sklearn.linear_model import LinearRegression
import numpy as np

# Data sample
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 7, 11])

model = LinearRegression()
model.fit(X, y)

prediksi = model.predict(X)
print("Koefisien:", model.coef_)
print("Intercept:", model.intercept_)
print("Prediksi:", prediksi)
```

**3. Machine Learning (*Linear Regression dengan Plot*)**

---



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data sample
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 7, 11])

# Membuat dan melatih model Linear Regression
model = LinearRegression()
model.fit(X, y)

# Melakukan prediksi menggunakan model yang telah dilatih
prediksi = model.predict(X)

# Menampilkan Koefisien, Intercept, dan Prediksi
print("Koefisien:", model.coef_)
print("Intercept:", model.intercept_)
print("Prediksi:", prediksi)

# Plot data asli
plt.scatter(X, y, color='blue', label='Data Asli')

# Plot garis regresi
plt.plot(X, prediksi, color='red', label='Garis Regresi')

# Menambahkan judul dan label
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

# Menampilkan plot
plt.show()
```

**3. Machine Learning (*K-Means Clustering*)**

---



```python
# K-Means Clustering
from sklearn.cluster import KMeans

# Data sample
data = np.array([[1, 2], [1, 4], [1, 0],
                 [10, 2], [10, 4], [10, 0]])

kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

print("Cluster Centers:", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)
```

**3. Machine Learning (*K-Means Clustering dengan Plot*)**

---



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Data sample
data = np.array([[1, 2], [1, 4], [1, 0],
                 [10, 2], [10, 4], [10, 0]])

# Membuat dan melatih model K-Means
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

# Mendapatkan hasil clustering
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Menampilkan Cluster Centers dan Labels
print("Cluster Centers:", cluster_centers)
print("Labels:", labels)

# Plot data asli dan hasil clustering
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', label='Data Points')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=200, alpha=0.75, label='Cluster Centers')

# Menambahkan judul dan label
plt.title('K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Menampilkan plot
plt.show()
```

**5. Introduction to Data Science**

---



```python
import pandas as pd

# Data sample
data = {'Usia': [25, 30, 35, 40, 45],
        'Gaji': [50000, 60000, 70000, 80000, 90000]}

df = pd.DataFrame(data)

# Data Preprocessing: Normalisasi
df['Gaji_Normalisasi'] = (df['Gaji'] - df['Gaji'].min()) / (df['Gaji'].max() - df['Gaji'].min())

print(df)

# Exploratory Data Analysis: Statistik Deskriptif
print("Statistik Deskriptif:")
print(df.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data sample
data = {'Usia': [25, 30, 35, 40, 45],
        'Gaji': [50000, 60000, 70000, 80000, 90000]}

df = pd.DataFrame(data)

# Data Preprocessing: Normalisasi
df['Gaji_Normalisasi'] = (df['Gaji'] - df['Gaji'].min()) / (df['Gaji'].max() - df['Gaji'].min())

# Scatter Plot: Usia vs Gaji
plt.figure(figsize=(8, 5))
plt.scatter(df['Usia'], df['Gaji'], color='blue', label='Gaji')
plt.title('Scatter Plot Usia vs Gaji')
plt.xlabel('Usia')
plt.ylabel('Gaji')
plt.legend()
plt.grid(True)
plt.show()

# Line Plot: Usia vs Gaji
plt.figure(figsize=(8, 5))
plt.plot(df['Usia'], df['Gaji'], marker='o', linestyle='-', color='green', label='Gaji')
plt.title('Line Plot Usia vs Gaji')
plt.xlabel('Usia')
plt.ylabel('Gaji')
plt.legend()
plt.grid(True)
plt.show()

# Bar Plot: Usia vs Gaji_Normalisasi
plt.figure(figsize=(8, 5))
plt.bar(df['Usia'], df['Gaji_Normalisasi'], color='purple', label='Gaji Normalisasi')
plt.title('Bar Plot Usia vs Gaji Normalisasi')
plt.xlabel('Usia')
plt.ylabel('Gaji Normalisasi')
plt.legend()
plt.grid(True)
plt.show()
```
